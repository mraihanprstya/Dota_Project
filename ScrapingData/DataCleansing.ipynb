```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Rawdata/Raw_Data_afk.xlsx')
df
```

```python
dfID = pd.read_excel('Rawdata/EG_Match_ID.xlsx')
dfID
```

```python
#change name ID
for i in range(len(df)):
    temp1 = df['IDmatch'].iloc[i]
    if isinstance(temp1, str) == True:
        df['IDmatch'].iloc[i] = temp1[6:16]
for i in range(len(dfID)):
    temp1 = dfID['Match'].iloc[i]
    if isinstance(temp1, str) == True:
        dfID['Match'].iloc[i] = temp1[9:]
```

```python
#Remove duplicate and null on IDMatch
df = df.drop_duplicates(subset='IDmatch')
df = df.dropna(subset=['IDmatch'])
dfID = dfID.reset_index(drop=True)
```

```python
test = df.loc[df['Team'] == 'Evil Geniuses']
coba = df.loc[df['Team'] != 'Evil Geniuses']
coba = coba.set_axis(['IDmatch',
                      'Enemy',
                      'ECarry',
                      'P_ECarry',
                      'EMid',
                      'P_EMid',
                      'EOfflane',
                      'P_EOfflane',
                      'ESup',
                      'P_ESup',
                      'EHsup',
                      'P_EHsup',
                      'Team',
                      'Carry',
                      'P_Carry',
                      'Mid',
                      'P_Mid',
                      'Offlane',
                      'P_Offlane',
                      'Sup',
                      'P_Sup',
                      'Hsup',
                      'P_Hsup',
                      'win',
                      'Date'],axis=1)
df = pd.concat([test,coba],axis=0)
df = df.sort_values(by=['IDmatch'], ascending=False)
df = df.dropna()
df = df.reset_index(drop=True)
```

```python
df.to_excel('Rawdata/Semiraw_Data.xlsx', index = False)
```
